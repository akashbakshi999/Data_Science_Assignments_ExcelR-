```python
# Identify numerical columns
numerical_columns = df.select_dtypes(include=['number']).columns
print("Numerical Columns:", numerical_columns)
```

```python
# Calculating the mean, median, mode, and standard deviation for each numerical column
for column in numerical_columns:
    print(f"Statistics for {column}:")
    print(f"  Mean: {df[column].mean()}")
    print(f"  Median: {df[column].median()}")
    print(f"  Mode: {df[column].mode()[0]}")  # mode() returns a series, take the first value
    print(f"  Standard Deviation: {df[column].std()}")
    print("-" * 40)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 8))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)  # Adjust number of rows and columns for your columns
    sns.histplot(df[column], kde=True, bins=15)
    plt.title(f"Histogram of {column}")
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, palette="Set2")

# Title and Labels
plt.title('Boxplots of Numerical Variables')
plt.ylabel('Values')
plt.xticks(rotation=45)

plt.show()
```

```python
# Identified categorical columns
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
print("Categorical Columns:", categorical_columns)
```

```python
# Plot style
sns.set(style="whitegrid")

# Plot bar charts for categorical columns
for col in categorical_columns:
    plt.figure(figsize=(10, 8))
    sns.countplot(data=df, x=col, hue=col, palette="Set2", legend=False)
    plt.title(f'Frequency of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Display first few rows before standardization
print("Before Standardization:\n", df.head())
```

```python
# Initialize StandardScaler
scaler = StandardScaler()

# Applied standardization
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Display first few rows after standardization
print("After Standardization:\n", df.head())
```

```python
print("Original DataFrame:\n", df)
```

```python
# Applied one-hot encoding
df_encoded = pd.get_dummies(df, columns=['Brand'], drop_first=True)

print("One-Hot Encoded DataFrame:\n", df_encoded)
```
